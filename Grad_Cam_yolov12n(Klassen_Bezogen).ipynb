```python
# ============================================================
# YOLO Decision-based Grad-CAM (FIXED VERSION)
# ============================================================

import os, glob, cv2, random
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F

from ultralytics import YOLO
from pytorch_grad_cam import GradCAM, GradCAMPlusPlus
from pytorch_grad_cam.utils.image import show_cam_on_image
from IPython.display import Image, display


# ============================================================
# CONFIG
# ============================================================
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
torch.cuda.manual_seed(SEED)
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False

IMG_DIR   = "/content/yolo_Dataset/images/test"
MODEL_PT  = "/content/yolo12n.pt"
OUT_DIR   = "/content/gradcam"

IMG_SIZE  = 416
CONF_TH   = 0.25
NUM_SHOW  = 50

device = "cuda" if torch.cuda.is_available() else "cpu"
os.makedirs(OUT_DIR, exist_ok=True)


# ============================================================
# LOAD IMAGES
# ============================================================
image_paths = []
for ext in ("jpg", "png", "jpeg", "JPG", "PNG"):
    image_paths += glob.glob(os.path.join(IMG_DIR, f"*.{ext}"))

# 1. Immer sortieren!
image_paths = sorted(image_paths)

# 2. Deterministische Auswahl
rng = np.random.default_rng(SEED)
indices = rng.permutation(len(image_paths))
selected_images = [image_paths[i] for i in indices[:NUM_SHOW]]

print(f"✅ {len(image_paths)} images found")
print(f"🎯 Using {len(selected_images)} deterministic evaluation images")

# 3. Optional: speichern für UNet
with open("/content/eval_images.txt", "w") as f:
    for path in selected_images:
        f.write(path + "\n")


# ============================================================
# LOAD YOLO
# ============================================================
yolo = YOLO(MODEL_PT)
yolo.to(device)
yolo.model.eval()
print("Classes:", yolo.names)
print(f"Number of classes: {len(yolo.names)}")


# ============================================================
# FIX 1: PROPER TARGET LAYER SELECTION
# ============================================================
class YOLOForward(nn.Module):
    def __init__(self, yolo):
        super().__init__()
        self.model = yolo.model.model

        # Find the detection head - usually the last layer before predictions
        self.target_layers = self._find_detection_layers()
        print(f"🎯 Target layers found: {len(self.target_layers)}")
        for i, layer in enumerate(self.target_layers):
            print(f"   Layer {i}: {type(layer).__name__}")

    def _find_detection_layers(self):
        """Find the actual detection head layers"""
        candidates = []


        for i, layer in enumerate(self.model):
            layer_type = type(layer).__name__

            # Target C2f blocks or Conv2d in the neck/head region
            if layer_type in ['C2f', 'C3', 'SPPF', 'Conv']:
                candidates.append(layer)
                print(f"  Candidate {len(candidates)}: Layer {i} - {layer_type}")

            # Stop before the Detect head
            if layer_type == 'Detect':
                break

        # Return the last 2-3 feature layers (best for visualization)
        if len(candidates) >= 2:
            return candidates[-2:]  # Use last 2 layers
        return candidates[-1:] if candidates else [self.model[-2]]

    def forward(self, x):
        outputs = []
        for m in self.model:
            if getattr(m, "f", -1) != -1:
                if isinstance(m.f, int):
                    x = outputs[m.f]
                else:
                    x = [outputs[j] if j != -1 else x for j in m.f]
            x = m(x)
            outputs.append(x)
        return x


# ============================================================
# FIX 2: PROPER RESHAPE TRANSFORM
# ============================================================
def reshape_transform(tensor):
    """Handle different activation formats"""
    if isinstance(tensor, (tuple, list)):
        # Take the first valid tensor
        for t in tensor:
            if torch.is_tensor(t) and len(t.shape) == 4:
                return t
        tensor = tensor[0]

    if not torch.is_tensor(tensor):
        return tensor

    # Ensure 4D: [batch, channels, height, width]
    if len(tensor.shape) == 3:
        tensor = tensor.unsqueeze(0)

    return tensor


# ============================================================
# FIX 3: CORRECT PREDICTION PARSING
# ============================================================
def parse_yolo_output(out):

    if isinstance(out, (tuple, list)):
        for o in out:
            if torch.is_tensor(o):
                out = o
                break

    if not torch.is_tensor(out):
        return None

    # Remove batch dimension
    if out.ndim == 3:
        out = out[0]  # [4+nc, num_predictions]

    # Transpose to [num_predictions, 4+nc]
    if out.shape[0] < out.shape[1]:
        out = out.T

    return out


# ============================================================
# FIX 4: PROPER BBOX COORDINATE HANDLING
# ============================================================
def select_best_detection(preds, conf_th, img_size=640):
    """
    Select best detection and return properly normalized bbox
    """
    if preds is None or preds.ndim != 2 or preds.shape[0] == 0:
        return None


    # No objectness score in v8+, just class probabilities

    if preds.shape[1] > 4:
        cls_scores, cls_idx = preds[:, 4:].max(dim=1)
        score = cls_scores  # No objectness multiplication in v8
    else:
        return None

    # Find best detection
    best_idx = torch.argmax(score).item()
    if score[best_idx] < conf_th:
        return None

    # Get bbox in pixel coordinates
    bbox_pixels = preds[best_idx, :4].detach().cpu().numpy()

    # Normalize to [0, 1]
    bbox_norm = bbox_pixels / img_size

    return best_idx, cls_idx[best_idx].item(), score[best_idx].item(), bbox_norm


# ============================================================
# FIX 5: SIMPLIFIED TARGET CLASS
# ============================================================
class YOLODetectionTarget:
    """Simpler target focusing on the detection score"""
    def __init__(self, det_idx, class_idx):
        self.det_idx = det_idx
        self.class_idx = class_idx

    def __call__(self, model_output):
        preds = parse_yolo_output(model_output)
        if preds is None or self.det_idx >= len(preds):
            return torch.tensor(0.0)

        # Return the class score for the best detection
        if preds.shape[1] > 4:
            return preds[self.det_idx, 4 + self.class_idx]
        return torch.tensor(0.0)


# ============================================================
# FIX 6: MINIMAL HEATMAP PROCESSING
# ============================================================
def process_cam(cam, threshold=0.3):
    """
    Minimal processing to preserve gradient information
    """
    # Remove negative values
    cam = np.maximum(cam, 0)

    # Normalize
    if cam.max() > 0:
        cam = cam / cam.max()

    # Soft thresholding
    cam = np.where(cam < threshold, cam * 0.3, cam)

    # Very light smoothing
    cam = cv2.GaussianBlur(cam, (3, 3), 0)

    return cam


# ============================================================
# DRAW BBOX
# ============================================================
def draw_detection(img, bbox_norm, label, color=(0, 255, 0)):
    """Draw bounding box with normalized coordinates"""
    h, w = img.shape[:2]

    x_c, y_c, box_w, box_h = bbox_norm
    x1 = int((x_c - box_w / 2) * w)
    y1 = int((y_c - box_h / 2) * h)
    x2 = int((x_c + box_w / 2) * w)
    y2 = int((y_c + box_h / 2) * h)

    # Draw box
    cv2.rectangle(img, (x1, y1), (x2, y2), color, 3)

    # Draw label
    cv2.rectangle(img, (x1, y1 - 25), (x1 + 200, y1), color, -1)
    cv2.putText(img, label, (x1 + 5, y1 - 7),
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 0), 2)

    return img


# ============================================================
# INITIALIZE MODEL + GRAD-CAM
# ============================================================
model_cam = YOLOForward(yolo).to(device).eval()

# Use GradCAM++ for better localization
gradcam = GradCAMPlusPlus(
    model=model_cam,
    target_layers=model_cam.target_layers,
    reshape_transform=reshape_transform,
)


# ============================================================
# IMAGE LOADER
# ============================================================
def load_image(path, size=640):
    bgr = cv2.imread(path)
    if bgr is None:
        raise ValueError(f"Cannot read: {path}")

    rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)
    rgb = cv2.resize(rgb, (size, size))
    rgb_float = rgb.astype(np.float32) / 255.0

    return rgb, rgb_float


# ============================================================
# MAIN GRAD-CAM FUNCTION
# ============================================================
def gradcam_on_image(img_path):
    try:
        rgb, rgb_f = load_image(img_path, IMG_SIZE)
    except Exception as e:
        print(f"❌ Error loading {img_path}: {e}")
        return None

    # Prepare input
    inp = torch.from_numpy(rgb_f).permute(2, 0, 1).unsqueeze(0).to(device)
    inp.requires_grad_(True)

    # Forward pass
    with torch.set_grad_enabled(True):
        out = model_cam(inp)
        preds = parse_yolo_output(out)

    # Find best detection
    sel = select_best_detection(preds, CONF_TH, IMG_SIZE)

    if preds is None or preds.ndim != 2 or preds.shape[0] == 0:
        print(f"⚠️  No predictions at all in {os.path.basename(img_path)}")
        return None

    sel = select_best_detection(preds, CONF_TH, IMG_SIZE)

    if preds is None or preds.ndim != 2 or preds.shape[0] == 0:
        print(f"⚠️  No predictions at all in {os.path.basename(img_path)}")
        return None

    if sel is None:
        # Keine gültige Detektion → trotzdem Top-1 wählen
        cls_scores, cls_idx = preds[:, 4:].max(dim=1)
        det_idx = torch.argmax(cls_scores).item()
        score = cls_scores[det_idx].item()
        class_idx = cls_idx[det_idx].item()

        bbox_pixels = preds[det_idx, :4].detach().cpu().numpy()
        bbox_norm = bbox_pixels / IMG_SIZE

        valid_detection = False
        print(f"⚠️  Below threshold detection: {score:.2f}")
    else:
        det_idx, class_idx, score, bbox_norm = sel
        valid_detection = True

    class_name = yolo.names[class_idx]

    print(f"✅ {os.path.basename(img_path)}: {class_name} ({score:.2f})")
    print(f"   BBox (normalized): center=({bbox_norm[0]:.3f}, {bbox_norm[1]:.3f}), "
          f"size=({bbox_norm[2]:.3f}, {bbox_norm[3]:.3f})")

    # Generate Grad-CAM
    target = [YOLODetectionTarget(det_idx, class_idx)]
    cam = gradcam(input_tensor=inp, targets=target)[0]

    # Process with minimal filtering
    cam = process_cam(cam, threshold=0.2)

    # Create visualization
    visualization = show_cam_on_image(rgb_f, cam, use_rgb=True)

    # Draw detection box
    if valid_detection:
        label = f"{class_name} {score:.2f}"
        visualization = draw_detection(visualization, bbox_norm, label)
    else:
        cv2.putText(
            visualization,
            f"{class_name} not Detected",
            (20, 40),
            cv2.FONT_HERSHEY_SIMPLEX,
            1.0,
            (255, 0, 0),
            2
        )

    # Save
    out_path = os.path.join(OUT_DIR, "fixed_" + os.path.basename(img_path))
    cv2.imwrite(out_path, cv2.cvtColor(visualization, cv2.COLOR_RGB2BGR))

    return out_path


# ============================================================
# RUN
# ============================================================
print(f"\n🚀 Processing {NUM_SHOW} images...\n")

for i, img in enumerate(selected_images):
    print(f"[{i+1}/{NUM_SHOW}] {os.path.basename(img)}")
    out = gradcam_on_image(img)
    if out:
        display(Image(filename=out))
    print()

print(f"\n✅ Done! Outputs in: {OUT_DIR}")
```
